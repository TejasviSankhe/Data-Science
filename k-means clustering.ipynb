### transform the data to fir our needs. Here we are trying to analyse the data for age,sex, state of persons using/not using internet on their handheld devices and if they access dating sites/apps using it.

```r
data<-pew[pew$intmob==1|pew$intmob==2,c('age','state','sex','intmob','date1a')]
data<-data[data$date1a==2|data$date1a==1,]
data<-na.omit(data)
data <- scale(data)
data<-data[1:1000,]
```

## Perform k-means using k=3, and plot the data with 2 principal components which were derived from the data. The plot here captures almost half of the information about the multivariate data

```r
km3 <- kmeans(data, 3)
clusplot(data, km3$cluster, stand=TRUE, color=TRUE, shade=TRUE, labels=4, lines=2, plotchar=TRUE,span=TRUE,
         col.p="dark blue", cex=0.8)
```

```r
km5 <- kmeans(data, 5)
clusplot(data, km5$cluster, stand=TRUE, color=TRUE, shade=TRUE, labels=4, lines=2, plotchar=TRUE,span=TRUE,
         col.p="dark blue", cex=0.8)
```

```r
df3<-data.frame("3",km3$tot.withinss,km3$betweenss,mean(km3$withinss))
df5<-data.frame("5",km5$tot.withinss,km5$betweenss,mean(km5$withinss))
names(df3)<-c("k","total_withinss","betweenss","mean_withinss")
names(df5)<-c("k","total_withinss","betweenss","mean_withinss")

attributes<-rbind(df3,df5)
```

```r
attributes
```

## Above are the various attributes for the two cluster models. As it can be observed 5 cluster model is a better fit than 3. 

### within_ss: This represents the distance of each point from its cluster and the value is smaller for k=5 as compared to k=3
### between_ss: This represents the distance between the cluster centres. Hence for k=5, centroids are far apart from each other as compared to k=3.
### mean withinss: Represents the gain for the models for various k's, for k=5 it falls by huge difference indicating it would be a better choice to k=3.
