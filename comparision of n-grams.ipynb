```r
library(ggplot2)
```

```r
twograms<- data.frame(Size_in_MB=c(0.3,0.85,1.5,3,6,8,11),Time_in_Seconds=c(7.08,7.82,7.6,8.15,9.4,9.48,10.62))
threegrams <- data.frame(Size_in_MB=c(0.3,0.85,1.5,3,6,8,11),Time_in_Seconds=c(6.99,8.17,7.99,9.92,13.3,13.5,15.6))
```

```r
ggplot(twograms, aes(Size_in_MB)) +                    
  geom_line(aes(y=twograms$Time_in_Seconds), colour="red",size=1 ) +  
  geom_line(aes(y=threegrams$Time_in_Seconds), colour="blue", size=1) +
  labs(x="Size in MB",y="Time in Seconds") 
```

## the plot shows the time taken in seconds to execute the map reduce job for different sizes of total input.
### Red line shows performance to 2grams and blue line shows performance for 3grams.
